```python
# -*- coding: utf-8 -*-
import pandas as pd
import json
```

```python
lines = [line.rstrip('\n') for line in open('price_dataeth.json')]
```

```python
df=pd.DataFrame()
```

```python
row = -1
for line in lines:
    row = row + 1
    data = json.loads(line)
    for key in data["result"][0].keys():
        df.loc[row,key] = data["result"][0][key]
```
